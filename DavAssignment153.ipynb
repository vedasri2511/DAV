```python
import pandas as pd

# Load the data
df=pd.read_csv('student_data.csv')
# Display column names
print("Columns in the dataset:")
print(df.columns.tolist())
```

```python
data_sorted = df.sort_values(by='G3', ascending=False)
print("\nData sorted by Final Grades (G3):")
print(data_sorted[['school', 'age', 'G3']].head())
```

```python
high_scorers = df[df['G3'] > 15]
print("\nStudents with Final Grade (G3) > 15:")
print(high_scorers[['school', 'age', 'G3']])
```

```python
def calculate_grade(g3_score):
    if g3_score >= 18:
        return 'A'
    elif g3_score >= 15:
        return 'B'
    elif g3_score >= 10:
        return 'C'
    elif g3_score >= 5:
        return 'D'
    else:
        return 'F'

df['Grade'] = df['G3'].apply(calculate_grade)
print("\nData with Grades based on G3:")
print(df[['G3', 'Grade']].head())
```

```python
school_group = df.groupby('school')['G3'].mean().reset_index()
print("\nAverage Final Grade (G3) by School:")
print(school_group)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Make sure you are using 'df' (your DataFrame)
# Example: df = pd.read_csv('your_file.csv')

plt.figure(figsize=(14, 6))

# Bar plot of average G3 by school
plt.subplot(1, 2, 1)
sns.barplot(x='school', y='G3', data=df, errorbar=None)  # <-- added data=df and errorbar=None
plt.title('Average Final Grade by School')

# Pie chart of Grade distribution
plt.subplot(1, 2, 2)
df['Grade'].value_counts().plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette('pastel')
)
plt.title('Grade Distribution')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

```python
#NUMPY OPERATIONS:
import numpy as np

# Convert your pandas DataFrame to a NumPy array
data_array = df.values

print("Shape of data:", data_array.shape)
print("First 5 rows of data (as NumPy array):\n", data_array[:5])
```

```python
# Extract G1, G2, G3 columns only
marks_array = df[['G1', 'G2', 'G3']].values

print("Shape of marks array:", marks_array.shape)
print("First 5 students' marks:\n", marks_array[:5])
```

```python
g3 = df['G3'].values

mean_g3 = np.mean(g3)
median_g3 = np.median(g3)
std_g3 = np.std(g3)

print(f"Mean of Final Grades (G3): {mean_g3:.2f}")
print(f"Median of Final Grades (G3): {median_g3}")
print(f"Standard Deviation of Final Grades (G3): {std_g3:.2f}")
```

```python
above_avg_indices = np.where(g3 > mean_g3)[0]

print("Indexes of students scoring above average in G3:", above_avg_indices)
print("Number of students scoring above average:", len(above_avg_indices))
```

```python
g3_min = np.min(g3)
g3_max = np.max(g3)

g3_normalized = (g3 - g3_min) / (g3_max - g3_min)

print("First 5 normalized G3 scores:\n", g3_normalized[:5])
```

```python
absences = df['absences'].values

total_absences = np.sum(absences)

print(f"Total absences of all students: {total_absences}")
```

```python
correlation_matrix = np.corrcoef(marks_array.T)

print("Correlation matrix between G1, G2, and G3:\n", correlation_matrix)
```

```python
max_g3 = np.max(g3)
min_g3 = np.min(g3)

print(f"Maximum Final Grade (G3): {max_g3}")
print(f"Minimum Final Grade (G3): {min_g3}")
```
